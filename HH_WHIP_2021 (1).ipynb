```python
hard_hits = []
counter = 1
for i in ids:

    statcast_data = statcast_pitcher('2021-04-01', '2021-10-03', player_id = i)
    
    high_ev = statcast_data[statcast_data['launch_speed'] > 95.0]

    hits = ['home_run','double', 'single', 'triple']

    hh = high_ev[high_ev['events'].isin(hits)]

    hard_hits.append(hh.shape[0])
    print(counter)
    counter += 1
```
